```python
import joblib
import pandas as pd

model = joblib.load("sleep_quality_model.pkl")
```

```python
FEATURES = [
    "sleep_duration",        # minutes
    "deep_sleep_ratio",      # 0–1
    "rem_sleep_ratio",       # 0–1
    "hrv_rmssd",             # ms
    "resting_heart_rate",    # bpm
    "pm2_5",                 # µg/m³
    "noise_level",           # dB
    "temperature",           # °C
    "sleep_pressure",        # 0–100
    "steps"                  # count
]
```

```python
sample_input = {
    "sleep_duration": 420,        # 7 hours
    "deep_sleep_ratio": 0.22,     # 22%
    "rem_sleep_ratio": 0.24,      # 24%
    "hrv_rmssd": 45,              # moderate HRV
    "resting_heart_rate": 60,     # healthy RHR
    "pm2_5": 38,                  # moderate pollution
    "noise_level": 55,            # urban night noise
    "temperature": 26,            # room temperature
    "sleep_pressure": 68,         # elevated fatigue
    "steps": 8500                 # active day
}
```

```python
X_test = pd.DataFrame([sample_input], columns=FEATURES)
```

```python
prediction = model.predict(X_test)
print("Predicted Sleep Quality:", round(prediction[0], 2))
```
